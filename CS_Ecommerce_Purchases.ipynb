 How many people have a credit card that expires in 2020?

```python
df.columns
```

```python
df["CC Exp Date"]
```

```python
def func():
  count = 0
  for date in df["CC Exp Date"]:
    if date.split("/")[1] == "20":
      count = count+1
  print(count)
func()
```

```python
df["CC Exp Date"].apply(lambda x : x[3:] == "20")
```
